```python
import pandas as pd
df = pd.DataFrame(results, columns=['Estimator', 'Vectorizer', 'ROC-AUC', 'mAP'])
df['Estimator'].replace({'mnb': 'Multinomial Naive Bayes',
                         'bnb': 'Bernoulli Naive Bayes',
                         'cnb': 'Complement Naive Bayes',
                         'perceptron': 'Perceptron',
                         'svm': 'Linear SVM',
                         'gboost': 'Gradient Boosting Machine',
                         'logreg': 'Logistic Regression',
                         'rf': 'Random Forest',
                         },
                        inplace=True)
df.head()
```

```python
import seaborn as sns
plt.rcParams.update({'font.size': 20})
plt.figure(figsize=(16, 10))
sns.scatterplot(x='ROC-AUC',
                y='mAP',
                hue='Estimator',
                data=df,
                s=100,
                palette=['#FF0000',
                         '#00FF00',
                         '#0000FF',
                         '#000000',
                         '#AF00AF',
                         '#FF3F00',
                         '#00FFFF',
                         '#30A000',
                         ])
plt.title('Estimator performance on both mAP and AUC metrics')
plt.legend(loc='best')
plt.savefig('c:/git/spring2020-ml-project/graphs/map_auc.png')
```
